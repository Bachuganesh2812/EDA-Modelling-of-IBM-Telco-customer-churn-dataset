```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
demographics=pd.read_excel("/kaggle/input/telco-customer-churn-1113/Telco_customer_churn_demographics.xlsx")
demographics.head()
```

```python
location=pd.read_excel("/kaggle/input/telco-customer-churn-1113/Telco_customer_churn_location.xlsx")
location.head()
```

```python
population=pd.read_excel("/kaggle/input/telco-customer-churn-1113/Telco_customer_churn_population.xlsx")
population.head()
```

```python
services=pd.read_excel("/kaggle/input/telco-customer-churn-1113/Telco_customer_churn_services.xlsx")
services.head()
```

```python
status=pd.read_excel("/kaggle/input/telco-customer-churn-1113/Telco_customer_churn_status.xlsx")
status.head()
```

```python
# merging the demographics table with locations table on customer id

m1=pd.merge(demographics,location, how='outer', on='Customer ID')
```

```python
m1
```

```python
# merging the m1 table with services table on customer id

m2=pd.merge(m1,services, how='outer', on='Customer ID')
```

```python
m2
```

```python
# droping the column names count_x from the m2 table as it is not needed and 
# the status table also has a comun named count_x which will create a problem in merging 

m2.columns
m2.drop(['Count_x'],axis=1,inplace=True)
```

```python
# merging the m2 table with status table on customer id

m3=pd.merge(m2,status, how='outer', on='Customer ID')
```

```python
m3
```

```python
# merging the m3 table with population table on zipcode, this relation ships is many to many

m4=pd.merge(m3,population,how='outer',on='Zip Code')
```

```python
m4
```

```python
m4.isnull().sum()
```

```python
m4.shape
```

```python
m4.to_excel('merging all the tables.xlsx')
```

```python
m4.head()
```

```python
# closer inspection os some columns we can see a lot of null values which can
# imputed with the coresponding data

m4['Offer'].fillna('No Offer',inplace=True)
m4['Internet Type'].fillna('No Internet',inplace=True)
m4['Churn Category'].fillna('Continues With Us',inplace=True)
m4['Churn Reason'].fillna('Did Not Leave',inplace=True)
```

```python
# showing the null values in the new dataframe after imputations

m4.isnull().sum()
```

```python
# dropping the null values

m4.dropna(inplace=True)
```

```python
# after droping the null values

m4.isnull().sum()
```

```python
# dropping some of the columns which have almost the same information but 
# different column names

m4.drop(['Customer ID','Under 30','Senior Citizen','Dependents','Count_y','Country','State','Lat Long','Latitude','Longitude','Count_x','Quarter_x','Referred a Friend','Count_y','Quarter_y','ID'],axis=1,inplace=True)
```

```python
m4.columns
```

```python
# dropping the less important columns
# and
# dropping some other columns which will directly tell us about the churn

m4.drop(['City','Zip Code','Customer Status','Churn Label','Churn Score', 'CLTV', 'Churn Category', 'Churn Reason','Population'],axis=1,inplace=True)
```

```python
m4.columns
```

```python
m4.info()
```

```python
m4.to_excel('after_preprocessing.xlsx')
```

```python
# selecting the columns which need to be converted into numerical type

label_encode_cols=m4.select_dtypes(include=["object"]).columns
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for col in label_encode_cols:
    m4[col] = label_encoder.fit_transform(m4[col])
```
